```python
import pandas as pd
import matplotlib.pyplot as plt

df_org = pd.read_csv("../../data\\raw\\cars_contactcars.csv")
df_detailed = pd.read_csv("../../data\\raw\\detailed_urls_data.csv")

df_detailed.head(15)
```

```python
df_main = pd.concat([df_org, df_detailed], axis=1)
df_main.shape
```

```python
df_main.drop(columns=["Details_URL", "URL"], inplace=True)
df_main.head()
```

```python
df_main.to_csv("../../data/raw/original_dataset.csv", encoding="utf-8-sig", index=False)
```

```python
# Feature Engineering
import re

years = df_main['Model_name'].tolist()
modified_years = []

for year in years:
    year_only = re.search(r'\b(19|20)\d{2}\b', year)
    modified_years.append(year_only.group())

df_main['Year_model'] = pd.Series(modified_years)
df_main.head()
```

```python
list_of_car_names = df_main['Year_model'] + " " + df_main['Manufacturer']
list_of_model_names = df_main['Model_name']
differences = []

for i in range(len(list_of_car_names)):
    words1 = list_of_car_names[i].split()
    words2 = list_of_model_names[i].split()

    diff = [word for word in words2 if word not in words1]
    res = " ".join(diff)
    differences.append(res)

df_main["Car_name"] = pd.Series(differences)
df_main.head()
```

```python
new_order = ["Model_name", "Year_model", "Manufacturer","Car_name", "Type", "Condition","Distance", "Color", "Body_type", "Fuel_type", "Engine_cc", "Price"]

df_main_ordered = df_main.reindex(columns=new_order)

df_main_ordered.head()
```

```python
df_main_ordered["Distance"] = df_main_ordered["Distance"].fillna("0 كم")

df_main_ordered["Distance"] = df_main_ordered["Distance"].str.replace(",", "")
df_main_ordered["Distance"] = df_main_ordered["Distance"].str.replace("كم", "").astype("int64")

df_main_ordered["Distance"].dtype

```

```python
df_main_ordered = df_main_ordered.dropna()
df_main_ordered.isna().sum()
```

```python
with open("../../data/lists/car_brands.txt", 'a', encoding='utf-8-sig') as f:
    for brand in df_main_ordered.sort_values(by="Manufacturer")["Manufacturer"].unique():
        f.write(brand + "\n")
```

```python
from pathlib import Path
import os
import re

data_list = Path(os.getcwd()).resolve().parents[1] / "data" / "lists"
data_list.mkdir(parents=True, exist_ok=True)

dict_car_names = (
    df_main_ordered.groupby("Manufacturer")["Car_name"]
    .agg(list)
    .to_dict()
)

dict_car_types = (
    df_main_ordered
    .groupby(["Manufacturer", "Car_name"])["Body_type"]
    .unique()
    .to_dict()
)

def clean_name(name):
    return re.sub(r'\s+', ' ', str(name).strip())

for manufacturer in dict_car_names:

    manufacturer_dir = data_list / clean_name(manufacturer)
    manufacturer_dir.mkdir(exist_ok=True)

    file_path = manufacturer_dir / f"{clean_name(manufacturer)}.txt"

    with open(file_path, "w", encoding="utf-8-sig") as f:
        for car_name in sorted(set(dict_car_names[manufacturer])):

            body_types = dict_car_types.get(
                (manufacturer, car_name), []
            )

            body_types = sorted(set(map(str, body_types)))

            if body_types:
                f.write(f"{car_name} : {', '.join(body_types)}\n")
            else:
                f.write(f"{car_name} : Unknown\n")

print("Done")
```
